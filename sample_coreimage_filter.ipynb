画像にさまざまな処理（フィルター）を掛ける

```python
from PIL import Image
from matplotlib import pylab as plt
import numpy as np

pil_img = Image.open("Mona_Lisa.jpg")
np_img = np.array(pil_img)

plt.figure(figsize=[6,6])
plt.imshow(np_img, aspect=1.0)
```

```python
from uikit.ui_uiimage_convert import *
from coreimage.core_image import *

np_img2 = apply_filter2np(np_img, 'CICMYKHalftone', {})

plt.figure(figsize=[6,6])
plt.imshow(np_img2, aspect=1.0)
```

```python
np_img2 = apply_filter2np(np_img, 
            'CISepiaTone', {'inputIntensity':1})

plt.figure(figsize=[6,6])
plt.imshow(np_img2, aspect=1.0)
```
