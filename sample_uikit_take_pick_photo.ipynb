```python
# PILやnumpy, matplotlibを読み込む
from PIL import Image
import numpy as np
from matplotlib import pylab as plt

# ファイルから画像を読み込む
img = np.array( Image.open('_take.jpg') )
# 画像を表示する
plt.figure(figsize=[6,6])
plt.imshow( img )
```

```python
from PIL import ImageFilter

# 下記行が向けていたので修正
img  = Image.open('_take.jpg')

# ガウシアンボカしをかけてみる
blurred_img = img.filter(ImageFilter.GaussianBlur(radius=10))

# エッジ抽出をしてみる
edged_img = img.filter(ImageFilter.FIND_EDGES)

# 画像を表示する
plt.figure(figsize=[9,6])

# 1行、3列に区切り、画像を並べて表示する
plt.subplot(1,3,1)      # オリジナル画像 
plt.imshow(img)
plt.subplot(1,3,2)      # ガウシアンぼかし画像
plt.imshow(blurred_img)
plt.subplot(1,3,3)      # エッジ抽出画像
plt.imshow(edged_img)
```
