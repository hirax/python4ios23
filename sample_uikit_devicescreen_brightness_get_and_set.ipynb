周りの明るさを（画面の明るさを介して）取得する

```python
from uikit.ui_screen_brightness import *
import time

interval = 1
x = []
for j in range(30):
    i = get_screen_brightness()
    print(i)
    x.append(i)
    time.sleep(interval)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 横軸の時間リストを作成する
t = np.arange(0., 
              len(x)*interval,
              interval)
# 
plt.scatter(t, x, c="r"); plt.plot(t, x, c="r")

plt.title("Light Intensity Level", fontsize=20)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Relative Intensity', fontsize=16)
plt.tick_params(labelsize=12)
plt.grid()
plt.figure(figsize=[8,4])
plt.show()
```
