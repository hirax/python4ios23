高度（気圧）センサ値をチャートに図示する

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
# 高度（気圧）センサ機能を使うためのモジュールを読み込む
from coremotion.cm_altimeter import *
import time

# 高度（気圧）センサを使うためのAltimeterクラスのインスタンスを作る
my_altimeter = Altimeter()

# 高度（気圧）センサ値の取得開始 
my_altimeter.start_relative_altitude_updates()
# 10秒間計測を続ける
time.sleep(10) # 単位は秒
# 計測を終了させる
my_altimeter.stop_relative_altitude_updates()
```

```python
import matplotlib.pyplot as plt

timestamp = []; relativeAltitude = []; pressure = []

altitudes = my_altimeter.altitude_data
for a in altitudes:
    timestamp.append(a['timestamp'])
    relativeAltitude.append(a['relativeAltitude'])
    pressure.append(a['pressure'])

# 横軸を時刻・縦軸を気圧（’pressure’）をとした折れ線チャートを描く
plt.scatter(timestamp,relativeAltitude,c="r")
plt.plot(timestamp,relativeAltitude,c="r")
# 軸名や各種修飾をしておく
plt.title("Pressure data", fontsize=20)
plt.xlabel('Time (s)', fontsize=16)
plt.ylabel('Pressure (k Pascal)', fontsize=16)
plt.tick_params(labelsize=12)
plt.grid()
plt.figure(figsize=[8,4])
plt.show()
```

```python
plt.scatter(timestamp,pressure,c="r")
plt.plot(timestamp,pressure,c="r")

plt.title("Relative Altitude data", fontsize=20)
plt.xlabel('Time (s)')
plt.ylabel('Relative Altitude (m)')
plt.tick_params(labelsize=12)
plt.grid()
plt.figure(figsize=[8,4])
plt.show()
```
