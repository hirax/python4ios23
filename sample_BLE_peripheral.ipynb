# BLE でデータを送受信する

```python
from corebluetooth.core_bluetooth import  CBPeripheralManagerEx
import numpy as np
import math
import time

myCBPeripheralManager = CBPeripheralManagerEx()
```

```python
for t in np.linspace(0, 2*2*math.pi, 300):
    value = int(math.sin(t)*100)
    name = "SIG_{:03d}".format(value)
    myCBPeripheralManager.set_localName(name)
    #myCBPeripheralManager.add_service()
    #time.sleep(1)
    myCBPeripheralManager.startAdvertising()
    time.sleep(1)
    myCBPeripheralManager.stopAdvertising()
    time.sleep(1)
    #myCBPeripheralManager.removeAllServices()
    #time.sleep(0.5)
```

```python
from corebluetooth.core_bluetooth import  CBCentralManagerEx

data = []

def process(_dict):
    data.append(_dict)
    print(_dict)

myCBCentralManager = CBCentralManagerEx(process)
```

```python
import time

# サーチ開始
myCBCentralManager.scanForPeripherals()
# 一定時間サーチし続ける
time.sleep(60)
# サーチ終了
myCBCentralManager.stopScan()
```

```python
values = []
for datum in data:
    if 'LocalName' in datum is not None:
        name = str(datum["LocalName"])
        if name[0:4] == "SIG_":
            values.append( float(name[5:8]) )
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[8,4])
ax = fig.add_subplot(111)
plt.title("Data obtained from BLE", fontsize=20)
plt.xlabel("Time[au]", fontsize=16)
plt.ylabel("Amplitude", fontsize=16)
ax.grid()
ax.plot(values, color="#ff0000", alpha = 0.8)
```
