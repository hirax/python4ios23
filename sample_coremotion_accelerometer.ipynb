# 加速度センサ値を取得してチャートに図示する

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
# MotionManagerを使うためのモジュールを読み込む
from coremotion.cm_motion_manager import *

# MotionManagerクラスのインスタンスを作る
my_motion_manager = MotionManager()

# 加速度センサ値の更新時間(単位は秒)を指定した上で、取得開始 
my_motion_manager.start_accelerometer_updates(0.1)
# 10秒間計測を続ける
time.sleep(10) # 単位は秒
# 計測を終了させる
my_motion_manager.stop_accelerometer_updates()
```

```python
import matplotlib.pyplot as plt

data = my_motion_manager.accelerometer_data

# x,y,z 各方向に対する加速度、時間リスト
x = []; y = []; z = []; t = []
for a in data:
    x.append(a['x']); y.append(a['y'])
    z.append(a['z']); t.append(a['at'])

plt.scatter(t,x,c="r"); plt.plot(t,x,c="r")
plt.scatter(t,y,c="g"); plt.plot(t,y,c="g")
plt.scatter(t,z,c="b"); plt.plot(t,z,c="b")

plt.title("Accelerometer data", fontsize=20)
plt.xlabel('Time (s)', fontsize=16)
plt.ylabel('Acclerometer (m/s^2)', fontsize=16)
plt.tick_params(labelsize=12)
plt.grid()
plt.figure(figsize=[8,4])
plt.show()
```
