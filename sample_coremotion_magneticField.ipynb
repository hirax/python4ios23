# 磁気センサー等から3軸の磁気値を取得する

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
from coremotion.cm_motion_manager import *
import time

# センサなどを扱うMotionManagerクラスのインスタンスを作る
my_motion_manager = MotionManager()

# 計測時間間隔を設定する（単位：秒）
interval = 0.1
my_motion_manager.start_magnetic_field_updates(interval)
time.sleep(5)
my_motion_manager.stop_magnetic_field_updates()  # 計測終了
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = my_motion_manager.magnetic_field_data
x = [d['x'] for d in data]
y = [d['y'] for d in data]
z = [d['z'] for d in data]

# 横軸の時間リストを作成する
t = np.arange(0., 
              len(x)*interval,
              interval)

# x,y,z各方向の磁界センサー値を描く
plt.scatter(t, x, c="r"); plt.plot(t, x, c="r")
plt.scatter(t, y, c="g"); plt.plot(t, y, c="g")
plt.scatter(t, z, c="b"); plt.plot(t, z, c="b")

plt.title("Magnetic Field data", fontsize=20)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Magnetic Field ($\mu T$)', fontsize=16)
plt.tick_params(labelsize=12)
plt.grid()
plt.figure(figsize=[8,4])
plt.show()
```
